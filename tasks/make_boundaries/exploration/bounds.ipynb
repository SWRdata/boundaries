```python
import geopandas as gp
import shapely as shp
from matplotlib import pyplot as plt
from tqdm import tqdm
tqdm.pandas()
```

```python
data = {}
for k in ["LAN"]:
    data[k] = gp.read_file(f"zip://../tmp/raw/2025-01-01_vg250_01-01.utm32s.shape.ebenen.zip!vg250_01-01.utm32s.shape.ebenen/vg250_ebenen_0101/VG250_{k}.shp")
```

```python
def make_bounds(land: str):
    d = data["LAN"].loc[data["LAN"]["GEN"] == land].dissolve().to_crs("wgs84")
    print(d.bounds.values[0])

make_bounds("Baden-Württemberg")
make_bounds("Rheinland-Pfalz")
```
